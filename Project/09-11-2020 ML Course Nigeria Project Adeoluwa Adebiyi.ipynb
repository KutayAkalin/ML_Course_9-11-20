## Data Preprocessing

### Fix Data Imbalance using Upsampling method

```python
from sklearn.utils import resample
from sklearn.model_selection import train_test_split, cross_validate

y = data["Churn"]

X = data.drop("Churn",axis=1)

from sklearn.preprocessing import StandardScaler

X_Scaler = StandardScaler()

X_Scaler.fit(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

new_df = pd.concat([X_train, y_train], axis=1)
print(new_df.info())
sns.countplot(new_df["Churn"])

X
```

#### Split Non-Churn data from Churn data

```python
churns = new_df[new_df.Churn==1]
non_churns = new_df[new_df.Churn==0]
```

```python
print(churns.info())
print(non_churns.info())
```

#### Upsample Churn Dataset to match size of Non-Churn

```python
upsampled_churn = resample(churns,
                           n_samples=len(non_churns),
                           replace=True,
                           random_state=42)

down_sampled_non_churn = resample(non_churns,
                                  n_samples=len(churns),
                                  replace=True,
                                  random_state=42)
print("UpSampled Info:")
print(upsampled_churn.info())
print("")
print("")
print("DownSampled Info:")
print(down_sampled_non_churn.info())
```

#### Merge Upsampled & Downsampled datasets

```python
new_df = pd.concat([non_churns, upsampled_churn])
new_df.head()

d_new_df = pd.concat([down_sampled_non_churn, churns])
d_new_df.head()
```

```python
print("Merged Dataset Info:")
print(new_df.info())

print("")
print("")

print("Merged Down-Sampled Dataset Info:")
print(d_new_df.info())
```

```python
sns.countplot(new_df["Churn"])
```

```python
sns.countplot(d_new_df["Churn"])
```

```python
sns.distplot(new_df["AccountWeeks"])
```

```python
sns.distplot(d_new_df["AccountWeeks"])
```

#### Data scaling for Logistic Regression

```python
y = new_df["Churn"]
X = new_df.drop("Churn",axis=1)
X
```

```python
y_d = d_new_df["Churn"]
X_d = d_new_df.drop("Churn",axis=1)
X_d
```

```python
# from sklearn.preprocessing import StandardScaler

# X_Scaler = StandardScaler()
X_Scaled = X_Scaler.transform(X)
X_Scaled
```

```python
X_d_Scaled = X_Scaler.transform(X_d)
X_d_Scaled
```

```python
sns.distplot(X_Scaled)
```

```python
sns.distplot(X_d_Scaled)
```

### Models

#### Logistic Regression Models

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(C=1,random_state=42)
cv_log_res = cross_validate(model,X_Scaled,y,cv=3, n_jobs=-1, return_estimator=True)
print(f"Mean training_accuracy: {np.mean(cv_log_res['test_score'])}")
print(f"Test Accuracy: {cv_log_res['estimator'][0].score(X_Scaler.transform(X_test),y_test)}")
```

```python
from sklearn.metrics import classification_report
print("Classification Report:")

print(classification_report(y_test,cv_log_res['estimator'][0].predict(X_Scaler.transform(X_test))))


```

```python
from sklearn.metrics import roc_curve, auc

final_model = cv_log_res["estimator"][0]

y_pred_prop = final_model.predict_proba(X_Scaler.transform(X_test))[:,1]

fpr_log, tpr_log, _ = roc_curve(y_test, y_pred_prop)
roc_auc_log = auc(fpr_log, tpr_log)

sns.set_style("white")
plt.figure(figsize=(10, 7))
plt.plot(fpr_log, tpr_log, color='darkorange',
         label='ROC curve (area = %0.2f)' % roc_auc_log)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate',fontsize=18,labelpad =10)
plt.ylabel('True Positive Rate',fontsize=18)
plt.title('Receiver Operating Characteristic',fontsize=22).set_position([.5, 1.02])
plt.legend(loc="lower right",fontsize=13)
plt.show()
```

```python
pred = cv_log_res["estimator"][0].predict(X_Scaler.transform(X_test))
cm = confusion_matrix(y_test, pred)
plt.figure(figsize=(12, 8))
ax =sns.heatmap(cm, square=True, annot=True, cbar=False)
ax.xaxis.set_ticklabels(["False","True"], fontsize = 12)
ax.yaxis.set_ticklabels(["False","True"], fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```
